## Mag and Phase Spectrum

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img = cv.imread('lake.jpg',0)
f = np.fft.fft2(img)
fshift1 = np.fft.fftshift(f)
magnitude_spectrumA = 20*np.log(np.abs(fshift1))
plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrumA, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img=cv2.imread('lake.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
dft1 = np.fft.fft2(img)
dft_shift1 = np.fft.fftshift(dft1)
phase_spectrumA = np.angle(dft_shift1)

ax1 = plt.subplot(1,2,1)
ax1.imshow(img, cmap='gray')

ax2 = plt.subplot(1,2,2)
ax2.imshow(phase_spectrumA, cmap='gray')

plt.show()
```
